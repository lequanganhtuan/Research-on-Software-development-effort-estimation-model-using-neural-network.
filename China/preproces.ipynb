```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Đọc file ARFF
with open('china.arff', 'r') as file:
    lines = file.readlines()

# Tìm dòng bắt đầu dữ liệu
data_start = lines.index('@data\n') + 1
data_lines = lines[data_start:]

# Lấy danh sách thuộc tính
attributes = []
for line in lines:
    if line.startswith('@attribute'):
        attr_name = line.split()[1]
        attributes.append(attr_name)

# Chuyển dữ liệu sang định dạng CSV
csv_data = []
for line in data_lines:
    line = line.strip()
    if line:
        # Thay thế khoảng trắng liên tiếp bằng dấu phẩy
        line = re.sub(r'\s+', ',', line)
        csv_data.append(line)

# Ghi ra file CSV
with open('china.csv', 'w') as f:
    # Ghi header
    f.write(','.join(attributes) + '\n')
    # Ghi dữ liệu
    f.write('\n'.join(csv_data))

# Đọc file CSV
df = pd.read_csv('china.csv')

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    negative_mask = df[col] < 0
    if negative_mask.any():
        print(f"Cột {col} có {negative_mask.sum()} giá trị âm")
        mean_value = df[col][~negative_mask].mean()
        df.loc[negative_mask, col] = mean_value
        print(f"Đã thay thế giá trị âm ở cột {col} bằng giá trị trung bình: {mean_value:.2f}")

df['Resource'] = df['Resource'].astype(int)

# Cột DevType là categorical, chuyển thành số (NewDev: 1, Maint: 0)
df['DevType'] = df['DevType'].map({'NewDev': 1, 'Maint': 0}).astype(int)



# Lọc các cột số (loại trừ Resource và DevType vì là biến phân loại)
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
numeric_cols = [col for col in numeric_cols if col not in ['Resource', 'DevType']]

# Vẽ boxplot trước khi xử lý ngoại lai
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(4, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot của {col} (Trước xử lý)')
plt.tight_layout()
plt.show()

# Xử lý ngoại lai bằng phương pháp clipping (giới hạn giá trị)
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Clipping: giới hạn giá trị trong khoảng [lower_bound, upper_bound]
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
    print(f"Đã xử lý ngoại lai ở cột {col}: lower_bound={lower_bound:.2f}, upper_bound={upper_bound:.2f}")


# Vẽ boxplot sau khi xử lý ngoại lai
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(4, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot của {col} (Sau xử lý)')
plt.tight_layout()
plt.show()


corr_matrix = df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, center=0)
plt.title('Ma trận tương quan của các biến')
plt.show()

df = df.drop('Added', axis=1)
df = df.drop('Deleted', axis=1)

scaler = StandardScaler()
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

df.to_csv('china_cleaned.csv', index=False)


# Tóm tắt dữ liệu
print("Tóm tắt dữ liệu:")
print(f"Số dòng: {len(df)}")
print(f"Số cột: {len(df.columns)}")
print("\nThông tin kiểu dữ liệu:")
print(df.dtypes)
print("\nThống kê mô tả:")
print(df.describe())
print("\nSố giá trị thiếu:")
print(df.isnull().sum())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Đọc dữ liệu từ file CSV
df = pd.read_csv('china.csv')

# 1. Kiểm tra dữ liệu ban đầu
print("Thông tin dataset ban đầu:")
print(df.info())
print("\nSố giá trị thiếu trong mỗi cột:")
print(df.isnull().sum())

# 2. Chuyển cột DevType thành dạng số (one-hot encoding)
valid_dev_types = ['NewDev', 'Maint']
if not df['DevType'].isin(valid_dev_types).all():
    print("Cột DevType chứa giá trị không hợp lệ, thay thế bằng 'Maint'")
    df['DevType'] = df['DevType'].apply(lambda x: x if x in valid_dev_types else 'Maint')
df = pd.get_dummies(df, columns=['DevType'], prefix='DevType', dtype=float)

# 3. Xử lý giá trị bất thường
numeric_columns = ['AFP', 'Input', 'Output', 'Enquiry', 'File', 'Interface', 
                   'Added', 'Changed', 'Deleted', 'Resource', 'Duration', 'Effort']
for col in numeric_columns:
    # Thay thế giá trị âm bằng 0
    if (df[col] < 0).any():
        print(f"Cột {col} chứa giá trị âm, thay thế bằng 0")
        df[col] = df[col].clip(lower=0)
    # Điền giá trị NaN bằng median
    if df[col].isnull().any():
        print(f"Cột {col} chứa giá trị NaN, điền bằng median")
        df[col].fillna(df[col].median(), inplace=True)

# 4. Xử lý giá trị ngoại lai bằng IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print(f"Xử lý ngoại lai cho cột {column}: lower_bound={lower_bound}, upper_bound={upper_bound}")
    df[column] = df[col].clip(lower=lower_bound, upper=upper_bound)
    return df

for col in ['AFP', 'Effort', 'Duration']:
    df = remove_outliers(df, col)

# 5. Kiểm tra tính nhất quán của AFP
df['Total_FP'] = df['Input'] + df['Output'] + df['Enquiry'] + df['File'] + df['Interface']
inconsistent_rows = df[abs(df['AFP'] - df['Total_FP']) / df['AFP'] > 0.1]
if not inconsistent_rows.empty:
    print(f"Tìm thấy {len(inconsistent_rows)} hàng có AFP không nhất quán với tổng các thành phần")
    df.loc[inconsistent_rows.index, 'AFP'] = df.loc[inconsistent_rows.index, 'Total_FP']
df = df.drop(columns=['Total_FP'])

# 6. Kiểm tra skewness để chọn phương pháp chuẩn hóa
skewness = df[numeric_columns].skew()
print("\nSkewness của các cột số:")
print(skewness)

# Chọn phương pháp chuẩn hóa: Min-Max nếu có cột lệch mạnh, ngược lại Z-score
if any(abs(skewness) > 1):
    print("Dữ liệu có độ lệch lớn, sử dụng Min-Max Scaling")
    scaler = MinMaxScaler()
else:
    print("Dữ liệu không lệch mạnh, sử dụng Z-score Standardization")
    scaler = StandardScaler()

# Áp dụng chuẩn hóa cho các cột số
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

# 7. Chuyển đổi kiểu dữ liệu
for col in numeric_columns:
    df[col] = df[col].astype('float64')

# 8. Lưu dataset đã chuẩn hóa
df.to_csv('china_normalized.csv', index=False)
print("Dataset đã được chuẩn hóa và lưu vào 'china_normalized.csv'")

# In thông tin dataset sau khi xử lý
print("\nThông tin dataset sau khi chuẩn hóa:")
print(df.info())
print("\nMô tả thống kê của dataset sau khi chuẩn hóa:")
print(df.describe())
```
